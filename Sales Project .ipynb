```python
# importing python libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt# visualizing data
import seaborn as sns
```

```python
# importing csv file

df_sales = pd.read_csv('Diwali Sales Data.csv', encoding = 'latin1')
```

```python
#to know the rows and columns
df_sales.shape
```

```python
#checking the top 5 rows
df_sales.head()
```

```python
#information about the DataFrame,data types, memory usage, range index,
df_sales.info()
```

```python
#droping unwanted rows and columns
df_sales.drop(columns=["Status","unnamed1"],inplace = True)
```

```python
df_sales.shape
```

```python
#check for null values
df_sales.isnull().sum().sort_values(ascending = False)
```

```python
#initiating null values by taking column values
mode_value=df_sales['Amount'].mode()[0]
```

```python
#knowing null values
mode_value
```

```python
#filling null values
df_sales['Amount'].fillna(mode_value,inplace=True)
```

```python
#rechecking it for null values
df_sales.isnull().sum()
```

```python
#Changing the data types
df_sales['Amount'] = df_sales['Amount'].astype('int')
```

```python
#Rechecking
df_sales['Amount'].dtype
```

```python
#checking the columns of the data frame
df_sales.columns
```

```python
df_sales[["Age","Marital_Status","Orders","Amount"]].describe()
```

# Visualization


```python
## plotting a bar chart for gender vs total amount
dfFM = df_sales["Gender"].value_counts().plot(kind='bar', color = ['orange','lightcoral'],title = "Gender")
```

```python
## plotting a bar chart for gender vs total amount

ax = sns.countplot(x='Gender',data = df_sales)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
#Checking in the code
df_sales.groupby(['Gender'])["Amount"].sum()
```

From above graphs we can see that most of the buyers are females

```python
df_sales
```

# Age Analyse

```python

dfageplot = df_sales.groupby(['Gender','Marital_Status','Age Group'])['Amount'].sum()

dfageplot.plot(kind='bar',figsize = (10,5),color = ['g','y','red'])
```

```python
# Total Amount vs Age Group

femaleplot = df_sales[df_sales['Gender'].isin(['F'])].groupby(['Age Group'])['Amount'].sum()

femaleplot.plot(kind='pie', title = 'Female age group expenses')
```

From above graphs we can see that most of the buyers are of age group B/W 25-35 year female

# State

```python
# total amount of sales from states 

stateplot = df_sales.groupby(['State'])['Amount'].sum().sort_values(ascending = False)

stateplot.plot(kind='bar',color=['g','y'],title='State Wise Revenue',figsize=(10,4))
```

# Marital Status

```python
MaritalStatus = df_sales.groupby(["Marital_Status",'Gender'])['Amount'].sum().sort_values(ascending=False)
MaritalStatus.plot(kind='bar',  color = ['pink','green'],title='MaritalStatus')
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

# Occupation

```python
Occupationplot = df_sales.groupby(['Occupation'])["Amount"].sum().sort_values(ascending = False)
Occupationplot.plot(kind='pie',)
```

*From above chart we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

# Product Category


```python
ProductCategory = df_sales.groupby(["Product_Category"])['Amount'].sum().sort_values(ascending=False)
ProductCategory.plot(kind='bar',figsize=(10,4),color=['red','orange','g','gold','y'])
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
# top 10 most sold products IDS
ProductIDplot = df_sales.groupby(['Product_ID'])['Amount'].sum().sort_values(ascending = False).head(10)
ProductIDplot.plot(kind='bar')
```

# Zone

```python
Zoneplot = df_sales.groupby(["Zone"])['Amount'].sum().sort_values(ascending = False)
Zoneplot.plot(kind='line',title='Zoneplot')
```

*From above graphs we can see that most of the revenue is genrating from Central zone*
